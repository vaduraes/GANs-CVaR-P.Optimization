```python
def MMD(x, y, kernel="rbf"):
    """Emprical maximum mean discrepancy. The lower the result
       the more evidence that distributions are the same.

    Args:
        x: first sample, distribution P
        y: second sample, distribution Q
        kernel: kernel type such as "multiscale" or "rbf"
    """
    device = torch.device("cpu")
    x.to(device)
    y.to(device)

    xx, yy, zz = torch.mm(x, x.t()), torch.mm(y, y.t()), torch.mm(x, y.t())
    rx = (xx.diag().unsqueeze(0).expand_as(xx))
    ry = (yy.diag().unsqueeze(0).expand_as(yy))

    dxx = rx.t() + rx - 2. * xx 
    dyy = ry.t() + ry - 2. * yy 
    dxy = rx.t() + ry - 2. * zz 

    dxx=dxx.to(device)
    dyy=dyy.to(device)
    dxy=dxy.to(device)

    XX, YY, XY = (torch.zeros(xx.shape).to(device),
                  torch.zeros(xx.shape).to(device),
                  torch.zeros(xx.shape).to(device))

    if kernel == "multiscale":
        
        bandwidth_range = [0.2, 0.5, 0.9, 1.3]
        for a in bandwidth_range:
            XX += a**2 * (a**2 + dxx)**-1
            YY += a**2 * (a**2 + dyy)**-1
            XY += a**2 * (a**2 + dxy)**-1
            
    if kernel == "rbf":
      
        bandwidth_range = [10, 15, 20, 50]
        for a in bandwidth_range:
            XX += torch.exp(-0.5*dxx/a)
            YY += torch.exp(-0.5*dyy/a)
            XY += torch.exp(-0.5*dxy/a)
     
    return torch.mean(XX + YY - 2. * XY)
```

```python
#KID comparison Synthetic with Synthetic
KID_SS=[]
def generate_latent_points(latent_dim_G, n_samples):
  x_input = randn(latent_dim_G * n_samples)
  x_input = x_input.reshape(n_samples, latent_dim_G)

  return x_input

#GENERATOR = load_model("/home/victor-duraes/Projects/Ocean2/GANs/SaveStep/GENERATOR_W_V0"+str(200)+".hdf5")
latent_dim_G=GENERATOR.input_shape[1]
n_samples=365


from tqdm import tqdm
for i in tqdm(range(1000)):
    x_input=generate_latent_points(latent_dim_G, n_samples)
    y=(GENERATOR.predict(x_input))+min_E
    y[:,Idx_Exclude==0]=0
    y[y>1]=1
    y[y<0]=0
    y=y[:,ID_NotZero==1]

    x_input=generate_latent_points(latent_dim_G, n_samples)
    x=(GENERATOR.predict(x_input))+min_E
    x[:,Idx_Exclude==0]=0
    x[x>1]=1
    x[x<0]=0
    x=x[:,ID_NotZero==1]

    x_torch=torch.from_numpy(x).double()
    y_torch=torch.from_numpy(y).double()

    KID_SS.append(MMD(x_torch, y_torch, kernel="rbf"))
 
KID_SS=np.array(KID_SS)
```

```python
#KID comparison Real with Synthetic
KID_RS=[]
def generate_latent_points(latent_dim_G, n_samples):
  x_input = randn(latent_dim_G * n_samples)
  x_input = x_input.reshape(n_samples, latent_dim_G)

  return x_input

#GENERATOR = load_model("/home/victor-duraes/Projects/Ocean2/GANs/SaveStep/GENERATOR_W_V0"+str(200)+".hdf5")
# latent_dim_G=GENERATOR.input_shape[1]
n_samples=365
from tqdm import tqdm
for i in tqdm(range(1000)):
    x_input=generate_latent_points(latent_dim_G, n_samples)
    y=(GENERATOR.predict(x_input))+min_E
    y[:,Idx_Exclude==0]=0
    y[y>1]=1
    y[y<0]=0
    y=y[:,ID_NotZero==1]

    EnergyPu=EnergyPu[np.random.permutation(EnergyPu.shape[0]),:,:,:]
    x=EnergyPu[0:n_samples,:,:,:]
    x=x[:,ID_NotZero==1]

    x_torch=torch.from_numpy(x).double()
    y_torch=torch.from_numpy(y).double()

    KID_RS.append(MMD(x_torch, y_torch, kernel="rbf"))
 
KID_RS=np.array(KID_RS)
```

```python
#Load historical data
NPZ_Data=np.load("AggregatedData_Daily.npz", allow_pickle=True) 
EnergyPu=NPZ_Data["EnergyPu"]
Lat=NPZ_Data["Lat"]
Long=NPZ_Data["Long"]
EnergyPu=NPZ_Data["EnergyPu"]
StartYear=NPZ_Data["StartYear"]
EndYear=NPZ_Data["EndYear"]
WindEnergy_RatedPower=float(NPZ_Data["WindEnergy_RatedPower"])
WaveEnergy_RatedPower=float(NPZ_Data["WaveEnergy_RatedPower"])
OceanEnergy_RatedPower=float(NPZ_Data["OceanEnergy_RatedPower"])

NPZ_Data_Yearly=np.load("AggregatedData_Daily.npz", allow_pickle=True) 
EnergyPu_Yearly=NPZ_Data_Yearly["EnergyPu"]

Idx_Exclude=np.max(EnergyPu_Yearly[:,:,:,1],axis=0)<=0.001
EnergyPu[:,Idx_Exclude,:]=0

Idx_Exclude_Ocean=np.max(EnergyPu_Yearly[:,:,:,2],axis=0)<=0.05
EnergyPu[:,Idx_Exclude_Ocean,2]=0
    
#EnergyPu=np.round(EnergyPu*255)#Put data in image format for convenience
#Extend data 25x25> 26x26 to improve in the padding

Idx_Exclude=np.max(EnergyPu_Yearly[:,:,:,1],axis=0)>0.05

Idx_Exclude_Ocean=np.max(EnergyPu_Yearly[:,:,:,2],axis=0)>0.05
Idx_Exclude_Ocean=np.reshape(Idx_Exclude_Ocean*1,(25,25,1))

Idx_Exclude=np.reshape(Idx_Exclude*1,(25,25,1))
Idx_Exclude=np.repeat(Idx_Exclude,2,axis=2)

Idx_Exclude=np.concatenate((Idx_Exclude,Idx_Exclude_Ocean),axis=2)
Idx_Exclude=Idx_Exclude==1


min_E=np.mean(EnergyPu,axis=0)
MaxMin=1

def PlotData_Stage1(Scenario, Sites, Tech, Generation, SaveP="None", IdxInclude=Idx_Exclude):
    ShapeFileCoast="./GEO_data/ne_10m_coastline.shp"
    ShapeFileStates="./GEO_data/ne_10m_admin_1_states_provinces_lines.shp"
    X_LONG=np.zeros((int(len(Lat)*len(Long))),dtype=float)
    Y_LAT=np.zeros((int(len(Lat)*len(Long))),dtype=float)

    min_latitude=np.min(Lat)
    max_latitude=np.max(Lat)

    min_longitude=np.min(Long)
    max_longitude=np.max(Long)

    xlim =[min_longitude,max_longitude]
    ylim=[min_latitude, max_latitude]

    df = gpd.read_file(ShapeFileCoast)
    df1 = gpd.read_file(ShapeFileStates)

    fig, ax = plt.subplots(figsize  = None)
    df.plot(color='black',linewidth=1,ax=ax)
    df1.plot(color='black',linewidth=1,ax=ax)

    count=0

    for I_lat in range(len(Lat)):
        for I_long in range(len(Long)):
            X_LONG[count] = Long[I_long]
            Y_LAT[count]  = Lat[I_lat]

            count=count+1

    Markers=["^","s","D"]
    Lab=["1","2","3"]
    for i in range(len(Sites)):
        plt.scatter(X_LONG[Sites[i]],Y_LAT[Sites[i]], color="k",label=Lab[i], marker=Markers[i], facecolors='none',s=70,linewidths=2)

    if len(Sites)>0:
        plt.legend(framealpha=1)
        
    Include=IdxInclude[:,:,Tech].reshape(-1)
    Include=Include==1
    Eng=Generation[Scenario,:,:,Tech].reshape(-1)
    Eng=Eng[Include]
    plt.scatter(X_LONG[Include], Y_LAT[Include],c=Eng, s=5, cmap='jet')
    clb=plt.colorbar(format='%.2f')
    clb.ax.set_title('CF')
    ax.set_xlim(xlim)
    ax.set_ylim(ylim)

    if SaveP!="None":

        plt.savefig(SaveP, dpi=600, bbox_inches='tight')

        return 0


    return plt.show()
```

## Histograms

```python
tech=1
energy_R=EnergyPu
energy_S=Y_Generator
location=1

def HistPlot(tech, energy_R, energy_S,location, SavePath="Histograms"):
   
    Location_Wind=[[7,10],[11,20],[20,18]] #Lat/Long
    Location_Wave=[[4,4],[11,16],[20,18]]
    Location_OceanCurrent=[[1,12],[12,20],[5,17]]    
    
    if tech==0:
        location=Location_Wind[location]
    if tech==1:
        location=Location_Wave[location]
    if tech==2:
        location=Location_OceanCurrent[location]
        
    n_S, x_S, _ = plt.hist(energy_S[:,location[0],location[1],tech],bins=20,range=(0,1), density=True)
    bin_centers = 0.5*(x_S[1:]+x_S[:-1])
    plt.clf()
    
    plt.hist(energy_R[:,location[0],location[1],tech],bins=20,range=(0,1), density=True,label="PDF Real Data")
    plt.plot(bin_centers,n_S,c="r",label="PDF Synthetic Data") ## using bin_centers rather than edges
    
    
    plt.legend()
    plt.ylabel("Probability Density")
    plt.xlabel("Generation [pu]")
    plt.xlim(0,1)
    plt.savefig(SavePath, dpi=600, bbox_inches='tight')
    return plt.show()

HistPlot(tech, energy_R,energy_S, location)
```

## Compare two Different Sites - Vs Plots

```python
tech1=2
tech2=2
location1=1
location2=2

energy_R=EnergyPu
energy_S=Y_Generator

SavePath="VsPlots"

def VsPlot(tech1,tech2, location1,location2,energy_R, energy_S, SavePath="VsPlots"):
   
    Location_Wind=[[7,10],[11,20],[20,18]] #Lat/Long
    Location_Wave=[[4,4],[11,16],[20,18]]
    Location_OceanCurrent=[[1,12],[12,20],[5,17]]    
    NBins=20
    Name=["Wind","Wave","O.Current"]
    Loc1Idx=location1
    Loc2Idx=location2
    if tech1==0:
        location1=Location_Wind[location1]
    if tech1==1:
        location1=Location_Wave[location1]
    if tech1==2:
        location1=Location_OceanCurrent[location1]

    if tech2==0:
        location2=Location_Wind[location2]
    if tech2==1:
        location2=Location_Wave[location2]
    if tech2==2:
        location2=Location_OceanCurrent[location2]

    bin_centers=np.arange(1/NBins*0.5,1,1/NBins)

    Q95_S=[]
    Q50_S=[]
    Q5_S=[]

    Q95_R=[]
    Q50_R=[]
    Q5_R=[]

    Real1 =energy_R[:,location1[0],location1[1],tech1]
    Real2 =energy_R[:,location2[0],location2[1],tech2]
    Synth1=energy_S[:,location1[0],location1[1],tech1]
    Synth2=energy_S[:,location2[0],location2[1],tech2]

    for i in range(NBins):
        #Real
        if i<NBins-1:
            IdxReal1=(Real1>=(1/NBins*i))*(Real1<(1/NBins*(i+1)))
            R2=Real2[IdxReal1]
            Q5_R.append(np.quantile(R2,0.10))
            Q50_R.append(np.quantile(R2,0.5))
            Q95_R.append(np.quantile(R2,0.9))

            IdxSyn1=(Synth1>=(1/NBins*i))*(Synth1<(1/NBins*(i+1)))
            S2=Synth2[IdxSyn1]
            Q5_S.append(np.quantile(S2,0.10))
            Q50_S.append(np.quantile(S2,0.5))
            Q95_S.append(np.quantile(S2,0.9))

        if i==NBins-1:
            IdxReal1=(Real1>=(1/NBins*i))*(Real1<=(1/NBins*(i+1)))
            R2=Real2[IdxReal1]
            Q5_R.append(np.quantile(R2,0.10))
            Q50_R.append(np.quantile(R2,0.5))
            Q95_R.append(np.quantile(R2,0.9))

            IdxSyn1=(Synth1>=(1/NBins*i))*(Synth1<=(1/NBins*(i+1)))
            S2=Synth2[IdxSyn1]
            Q5_S.append(np.quantile(S2,0.10))
            Q50_S.append(np.quantile(S2,0.5))
            Q95_S.append(np.quantile(S2,0.9))

        #Synthetic



    plt.plot(bin_centers,Q5_R,c="blue",ls="--")
    plt.plot(bin_centers,Q50_R,c="blue")
    plt.plot(bin_centers,Q95_R,c="blue",ls="--")

    plt.plot(bin_centers,Q5_S,c="red",ls="--")
    plt.plot(bin_centers,Q50_S,c="red")
    plt.plot(bin_centers,Q95_S,c="red",ls="--")

    #plt.legend()
    plt.ylabel(Name[tech2]+" "+str(Loc2Idx+1)+" - Generation [pu]")
    plt.xlabel(Name[tech1]+" "+str(Loc1Idx+1)+" - Generation [pu]")
    plt.xlim(0,1)
    plt.ylim(0,1)
    plt.savefig(SavePath, dpi=600, bbox_inches='tight')

VsPlot(tech1,tech2, location1,location2,energy_R, energy_S, SavePath="VsPlots")
```

## Portfolio Pot Diversity

```python
tech=[0,1,2,0,0,1,1]
location=[0,1,2,0,0,0,1]
energy_R=EnergyPu
energy_S=Y_Generator


def HistPlot(tech, energy_R, energy_S,location, SavePath="Histograms"):
   
    Location_Wind=[[7,10],[11,20],[20,18]] #Lat/Long
    Location_Wave=[[4,4],[11,16],[20,18]]
    Location_OceanCurrent=[[1,12],[12,20],[5,17]]    
    E_S=np.zeros((energy_S.shape[0]))
    E_R=np.zeros((energy_R.shape[0]))
    
    for i in range(len(tech)):
        if tech[i]==0:
            L_Input=Location_Wind[location[i]]
            E_S=E_S+energy_S[:,L_Input[0],L_Input[1],tech[i]]/len(tech)
            E_R=E_R+energy_R[:,L_Input[0],L_Input[1],tech[i]]/len(tech)
            
        if tech[i]==1:
            L_Input=Location_Wave[location[i]]
            E_S=E_S+energy_S[:,L_Input[0],L_Input[1],tech[i]]/len(tech)
            E_R=E_R+energy_R[:,L_Input[0],L_Input[1],tech[i]]/len(tech)
            
        if tech[i]==2:
            L_Input=Location_OceanCurrent[location[i]]
            E_S=E_S+energy_S[:,L_Input[0],L_Input[1],tech[i]]/len(tech)
            E_R=E_R+energy_R[:,L_Input[0],L_Input[1],tech[i]]/len(tech)

    n_S, x_S, _ = plt.hist(E_S,bins=40, density=True)
    bin_centers = 0.5*(x_S[1:]+x_S[:-1])
    plt.clf()
    
    plt.hist(E_R,bins=40, density=True,label="PDF Real Data")
    plt.plot(bin_centers,n_S,c="r",label="PDF Synthetic Data") ## using bin_centers rather than edges
    
    
    plt.legend()
    plt.ylabel("Probability Density")
    plt.xlabel("Generation [pu]")
    plt.xlim(0,1)
    plt.savefig(SavePath, dpi=1200, bbox_inches='tight')
    return plt.show()

HistPlot(tech, energy_R,energy_S, location)
```
